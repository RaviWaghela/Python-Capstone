```python
#### data Cleaning
#Let's check if there are duplicate records
airbnb.duplicated().any()
```

```python
#Dropping columns that are unnecessary for our analysis
airbnb.drop('id', axis = 1, inplace= True)
airbnb.drop('host_name', axis = 1, inplace = True)
airbnb.drop('calculated_host_listings_count', axis=1, inplace=True)
airbnb.drop('last_review', axis = 1, inplace =True)

#Renaming columns
airbnb = airbnb.rename(columns={'neighbourhood_group': 'borough',
                            'neighbourhood': 'neighborhood',
                            'listing_price': 'price'})
airbnb.columns
```

```python
##Handling Missing Values

#counting null values for each value
missing_values_count = np.sum(airbnb.isnull())
missing_values_count
```

```python
total_cells = np.product(airbnb.shape)
total_missing =  missing_values_count.sum()

#calculating the percentage of missing values
percentage_missing = np.round((total_missing / total_cells) * 100, 2)
print("Percentage of missing values in the dataset:", percentage_missing) 
```

```python
#Imputation of missing values with median values
airbnb['reviews_per_month'] = airbnb.reviews_per_month.fillna(airbnb.reviews_per_month.median())
```

```python
#Dropping null values and counting the dropped rows
cleaned_df = airbnb.dropna(axis = 0, how ='any') 

#Dropping Rows with at least 1 null value
print(len(airbnb) - len(cleaned_df ), 'rows were dropped with missing values.') 
```

```python
#considering name column is insignificant number null values we can replace those with N/A
airbnb.fillna({'name':'N/A'},inplace = True)

#reviews per month provides information about the number of reviews per month, missing values means 0 reveiws for the corresponding last_review column
airbnb.fillna({'reviews_per_month':0,'last_review':0},inplace = True)
```

```python
airbnb.isnull().sum()
```

```python
#Excluding prices that are equal to zero
df = cleaned_df[cleaned_df['price'] != 0]
```

```python
df.head()
```

```python
#########Outliers
```

```python
# calculate interquartile range
q25, q75 = np.percentile(cleaned_df.price, 25), np.percentile(cleaned_df.price, 75)
iqr = q75 - q25
```

```python
#calculate the outlier cutoff
cut_off = iqr * 1.5
lower, upper = q25 - cut_off, q75 + cut_off
print('Percentiles: 25th=%.3f, 75th=%.3f, IQR=%.3f' % (q25, q75, iqr))
```

```python
# identify outliers
outliers = [x for x in cleaned_df.price if x < lower or x > upper]
print('Outliers observations:',len(outliers))
```

```python
#non outliers
non_outliers = [x for x in cleaned_df.price if x >= lower and x <= upper]
print('Non-outlier observations:', len(non_outliers))
```
